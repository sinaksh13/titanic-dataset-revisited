```python
data.isnull().sum() #Displays the number of null values present in each column; 
#had it been just data.isnull() it would have returned false if no null value and true if null value was present in the columns 
#the output shows that the column Age has 177 rows with null value, column Cabin has 687 rows with null value and so on
```

```python
data_final=data.drop(["Age","Cabin","Embarked","Sex","Name","Ticket"],axis=1)
#axis=0 means rows and axis=1 means columns
#we are dropping the columns which have null values or which are categorical in nature
#we use axis=1 to instruct the system that the variables we are dropping are columns and not rows
#had it been rows we would have used axis =0
```

```python
data_final.shape
#after removing 6 columns we are now left with 6 columns
```

```python
data_final.isnull().sum()
#no column now has null value
```

```python
x=data_final.drop(["Survived"],axis=1)
#in x we are storing the independent variables by dropping the column "Survived" which is the dependent variable
```

```python
y=data_final[["Survived"]]
#in y we are storing the dependent variable
```

```python
from sklearn.model_selection import train_test_split
# calling train_test_split method from class model_selection present in the package sklearn
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.10) 
#using the train_test_split method we are now splitting our data into test and training data
#10%data will be test data specified by test_size=0.10
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
#calling "LinearRegression" method from class "linear_model" present in the package "sklearn"
#calling "KNeighborsClassifier" method from class "neighbors" present in the package "sklearn"
#so on and so forth
```

```python
lr=LogisticRegression() 
knn=KNeighborsClassifier()
nb=GaussianNB()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
#creating an instance of function "LogisticRegression" in order to call it "lr" instead of "LogisticRegression" everytime i need to call Logistic Regression
#same for other functions as well
```

```python
lr.fit(X_train,y_train) 
knn.fit(X_train,y_train)
nb.fit(X_train,y_train)
dt.fit(X_train,y_train)
rf.fit(X_train,y_train)
#creating a linear regression model based on the training data by calling the instance lr
#creating a KNN model based on the training data by calling the instance knn
#creating a naive bayes model based on the training data by calling the instance nb
#creating a decision tree model based on the training data by calling the instance dt
#creating a random forrest model based on the training data by calling the instance rf
```

```python
p1=lr.predict(X_test)
p2=knn.predict(X_test)
p3=nb.predict(X_test)
p4=dt.predict(X_test)
p5=rf.predict(X_test)
#testing our linear regression model by pushing test data into our model
#same for all other models
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, precision_recall_fscore_support
#calling accuracy_score, precision_score, recall_score, precision_recall_fscore_support functions from class metrics present in the package sklearn
#this is done to choose one of the 5 models for our analysis
```

```python
print (accuracy_score(p1,y_test))
print (accuracy_score(p2,y_test))
print (accuracy_score(p3,y_test))
print (accuracy_score(p4,y_test))
print (accuracy_score(p5,y_test))
#displaying the accuracy score of each model
```

```python
print (precision_recall_fscore_support(p1,y_test))
print (precision_recall_fscore_support(p2,y_test))
print (precision_recall_fscore_support(p3,y_test))
print (precision_recall_fscore_support(p4,y_test))
print (precision_recall_fscore_support(p5,y_test))
#displaying f score of each model
#its better to use f score value to choose our model rather than accuracy , precision or recall score
#reason being accuracy score may be high due to overfitting of line or imbalance data
#f score includes both preecision as well as recall 
#higher the f score better the model
```

```python
#Note:
#we did not perform data scaling or engineering because of time restraint
#we just removed the columns with null values for our ease
#However,practically we should replace the null values with the mean, median or mode of the values present in that columns
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
probs = model.predict_proba(X_test)
```

```python
probs = probs[:, 1]
```

```python
auc = roc_auc_score(y_test, probs)
print('AUC: %.3f' % auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, probs)
```

```python
pyplot.plot([0, 1], [0, 1], linestyle='--')
```

```python
pyplot.plot(fpr, tpr, marker='.')
```
